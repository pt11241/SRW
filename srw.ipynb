```python
from srw2 import Quadratic_Irrationals, PellsEquations
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

If $\alpha \in \mathbb{R}$, then $\alpha$ is a quadratic irrational if and only if there exist $P, Q, D \in \mathbb{Z}$ such that $Q \neq 0$, $D > 0$ is not a perfect square, and $\alpha = P + \frac{\sqrt{D}}{Q}.$

```python
p,Q = 0, 1
```

```python
D = np.arange(1, 100000)
indx = np.sqrt(D) % 1 != 0
D = D[indx]
D.size
```

```python
lens = np.array([Quadratic_Irrationals(p, Q, i).period() for i in D])
```

Averaging 

```python
Lmed = Quadratic_Irrationals.averaging(lens)
```

Graph

```python
# df = pd.read_csv('output.csv')
# D = df['D'].to_numpy()
# Lmed = df['periods_averaging'].to_numpy()
# lens = df['lens_periods'].to_numpy()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

ax1.plot(D, Lmed, color='b', alpha=1, label=' ', linewidth=3)
ax1.set_title('  L(D)')
ax1.set_xlabel('D')
ax1.set_ylabel('L(D)')
ax1.grid()
ax1.legend()

ax2.plot(D, lens, 'o', color='g', alpha=0.1, label='')
ax2.set_title(' ')
ax2.set_xlabel('D')
ax2.set_ylabel('L(D)')
ax2.grid()
ax2.legend()

plt.tight_layout()
plt.show()
```
