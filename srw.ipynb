```python
from srw import Quadratic_Irrationals
import matplotlib.pyplot as plt
import numpy as np
```

If $\alpha \in \mathbb{R}$, then $\alpha$ is a quadratic irrational if and only if there exist $P, Q, D \in \mathbb{Z}$ such that $Q \neq 0$, $D > 0$ is not a perfect square, and $\alpha = P + \frac{\sqrt{D}}{Q}.$

```python
p,Q = 0, 1
```

```python
D = np.arange(1, 100000)
indx = np.sqrt(D) % 1 != 0
D = D[indx]
```

```python
lens = np.array([Quadratic_Irrationals(p, Q, i).period() for i in D])
```

Averaging 

```python
Lmed = Quadratic_Irrationals.averaging(lens)
```

Graph

```python
plt.figure(figsize=(14, 7))
DmedGraph = plt.plot(D, Lmed, color='b', alpha=1, label=' ', linewidth=3)
lens_D = plt.plot(D, lens, 'o', color='g', alpha=0.3, label='')
# plt.plot(D, Lmed, 'o-', color='b', alpha=0.8, label=' ')
# plt.plot(D, lens, '-', color='g', alpha=0.6, label='')
plt.legend()
plt.title("    ")
plt.grid()
plt.xlabel("D")
plt.ylabel("L(D)")
plt.show()
```
